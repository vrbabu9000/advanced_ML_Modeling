```python
#load the libraries
import pandas as pd
import numpy as np
import pandas_profiling as pp
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
```

## Import New Dataset

```python
raw_data = pd.read_csv("C:\\Users\\Vignesh R Babu\\excelR-datascience\\assignment_14_decision_trees\\14Q1\\Fraud_check.csv")
raw_data.rename(columns={'Undergrad':'undergrad','Marital.Status':'marital_status','Taxable.Income':'tax_income','City.Population':'city_pop','Work.Experience':'work_exp','Urban':'urban'}, inplace=True)
df = raw_data.copy()
df.head()
```

```python
#def func to assign fraud risky and good values

def filt(x):
    if x<=30000:
        return 'Risky'
    else:
        return 'Good'
```

```python
df['fraud'] = df['tax_income'].copy()
df['fraud'] = df['fraud'].apply(filt)
df.head()
```

```python
df.fraud.value_counts()
```

```python
label_encoder = preprocessing.LabelEncoder()
df['undergrad']= label_encoder.fit_transform(df['undergrad']) 
df['urban']= label_encoder.fit_transform(df['urban']) 
df['marital_status']= label_encoder.fit_transform(df['marital_status']) 
df['fraud']= label_encoder.fit_transform(df['fraud'])

df.head()
```

```python
df.info() # No null values
```

```python
df.describe()
```

## Visualization

```python
sns.pairplot(df,hue='fraud',palette='Dark2')
```

```python
sns.catplot(x='fraud',y='tax_income',data=df,kind='box',palette='Dark2')
```

## Feature Engineering

All the features in the dataset are relevant and can be used for model training.
There are no dummies in the dataset.

### Train and Test Split data

```python
X = df.iloc[:,:-1]
y =df.fraud
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

## Decision Tree Classifier
### Default Parameters

```python
model = DecisionTreeClassifier() 
```

```python
model.fit(X_train,y_train)
```

```python
base_pred = model.predict(X_test)
```

### Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
```

```python
confusion_matrix(y_test,base_pred) #This is very good. But not sure if it's an overfitted model
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
print(classification_report(y_test,base_pred))
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance']) 
#Only tax_income matters since our tree is based on it. 
```

### Visualizing the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8), dpi = 75)
plot_tree(model,filled=True,feature_names=X.columns);

#Note that criterion by default is Gini (CART)
```

## Reporting Model Results

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=75)
    plot_tree(model,filled=True,feature_names=X.columns);
```

## Experimenting with HyperParameters

```python
#Using max depth
trimmed_tree = DecisionTreeClassifier(max_depth=2)
trimmed_tree.fit(X_train,y_train)
```

```python
report_model(trimmed_tree) #We dont see any differenc since this model only has a depth of 2.
#doing this for future reference
```

```python
## Experimenting with max leaf nodes
trim2_tree = DecisionTreeClassifier(max_leaf_nodes=2)
trim2_tree.fit(X_train,y_train)
```

```python
report_model(trim2_tree) # we get the same result since this model only has a depth of 2
```

```python
#Experimenting with crtieria C.50
entropy_tree = DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree) #entropy values are visible
```
